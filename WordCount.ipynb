```
input_data = sc.textFile('/FileStore/tables/data.txt')
words = ["PARALLEL","PROCESSING","MULTIPROCESSING","MULTITHREADING","CPU","GPU"]
word_counts = input_data.flatMap(lambda line: line.split(" "))\
             .filter(lambda word: word if word.upper() in words else None)\
             .map(lambda word: (word.upper(), 1))\
             .reduceByKey(lambda a, b: a + b)
```

```
word_counts.collect()
```

```
input_data = sc.textFile('/FileStore/tables/data.txt')
pharse = "PARALLEL PROCESSING"
pharse_counts = input_data.flatMap(lambda line: line.split('.'))\
             .map(lambda line: (pharse,1) if pharse in line.upper() else (pharse,0))\
             .reduceByKey(lambda a, b: a + b)
```

```
pharse_counts.collect()
```
